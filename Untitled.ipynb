```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('rgb.png')
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
blurred = cv2.medianBlur(gray, 5)
```

```python
plt.imshow(blurred, cmap = "gray")
plt.show()
```

```python
minDist = 20  # Minimum distance between the centers of detected circles
param1 = 100   # Upper threshold for the internal Canny edge detector
param2 = 20   # Threshold for center detection
minRadius = 8  # Minimum radius of the circles to detect
maxRadius = 15  # Maximum radius of the circles to detect
```

```python
circles = cv2.HoughCircles(blurred, cv2.HOUGH_GRADIENT, 1, minDist, param1=param1, param2=param2, minRadius=minRadius, maxRadius=maxRadius)
```

```python
print(circles)
```

```python
if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0,:]:
        center = (i[0], i[1])
        radius = i[2]
        print(f"Center coordinates: {center}")
        
        # Draw the outer circle
        cv2.circle(img, center, radius, (0, 255, 0), 2)
        
        # Draw the center of the circle
        cv2.circle(img, center, 2, (0, 0, 255), 3)
        
        print(f"Center coordinates: {center}")
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

```python
# Display the image with the detected circles
cv2.imshow("Detected Circles", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
_, thresh = cv2.threshold(blurred, 220, 255, cv2.THRESH_BINARY)
```

```python
plt.imshow(thresh)
plt.show()
```

```python
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
min_area = 1000
max_area = 10000
sticker_contours = [c for c in contours if min_area < cv2.contourArea(c) < max_area]
```

```python
for c in sticker_contours:
    (x, y), radius = cv2.minEnclosingCircle(c)
    center = (int(x), int(y))
    radius = int(radius)
    cv2.circle(img, center, radius, (0, 255, 0), 2)
    cv2.circle(img, center, 2, (0, 0, 255), 3)
    print(f"{center[0]},{center[1]}")
```
